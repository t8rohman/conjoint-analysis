# Creating Choice Tables for Respondents to Answer

## Check the choice table first

The tables should be in format of:

- Having Questions ID (qID) column; indicating that one choice is a group of a set of choice.
- All attributes for each choice.
- Should be in format where every choices in each group of choices are put in their own respective row.

```python
import pandas as pd

# let's read the data from the docs
df = pd.read_csv('car_options_example.csv')
df
```

## Call the module and create the table

Parameters to pass:
- input_path = file that contains all the options for the question later. Should be in csv file.
- label = label name regarding the topic, for example cars, banks, computers.
- col_id = question ID, indicating that one choice is a group of a set of choice.
- path_to_save = Location to save the tables.

```python
from conjoint import table_generator

table_generator.table_generator(input_path='car_options_example.csv',
                                label='Cars',
                                col_id='qID',
                                path_to_save='images')
```
